Practical implementation of Simple RNN


```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
# define the vocabulary size 
voc_size = 10000
```

```python
#One hot representation
one_hot_repre = [one_hot(words, voc_size)for words in sent]
one_hot_repre
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length = 8
```

```python
import numpy as np
```

```python
# here we use padding sequences in order to make all sentences of same size.
embedded_docs = pad_sequences(one_hot_repre, padding='pre', maxlen=sent_length)
print(embedded_docs)
```

```python
## feature representation with dimension size of 10
dim = 10
```

```python
from tensorflow.keras.layers import Input
```

```python
model = Sequential()
# Move the shape definition here
model.add(Input(shape=(sent_length,))) 
# Remove input_length from here
model.add(Embedding(voc_size, dim)) 
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
